```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#If you want a comparision between features
#values are placeholder
results = pd.DataFrame({'model':['Constant','Type-OHE','LDP', 'Type+LDP','Properties'],
              'auroc-score':[0.5, 0.5, 0.5, 0.5,0.5]}).sort_values('auroc-score', ascending=False)
```

```python
ax = results.plot.bar(x='model', y='auroc-score', color='#ffcc99', ec='#ff8000',rot=0)
plt.title('KG Link Prediction RGCN models comparison')
#plt.ylim(0.6,1)
#plt.yticks(np.arange(0.6,1.05,0.05))
plt.grid()
#plt.savefig('kg_rgcn_linkpre_auroc_comparison.pdf', bbox_inches='tight')
plt.show()
```
